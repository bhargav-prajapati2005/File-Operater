```python
import os
from datetime import datetime

class JournalManager:
    def __init__(self, filename="journal.txt"):
        self.filename = filename

    def add_entry(self):
        """Appends a new entry to the file. Creates it if it doesn't exist."""
        try:
            # Mode 'a' (Append): Adds to the end of the file without overwriting.
            # It also creates the file if it does not exist.
            entry_text = input("Write your journal entry: ")
            timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            
            with open(self.filename, "a") as file:
                file.write(f"[{timestamp}]\n{entry_text}\n{'-'*30}\n")
            print("Entry saved successfully.")
        except PermissionError:
            print("Error: You do not have permission to write to this file.")
        except Exception as e:
            print(f"An unexpected error occurred: {e}")

    def view_entries(self):
        """Reads and displays all entries."""
        try:
            # Mode 'r' (Read): Opens the file for reading. 
            # Throws FileNotFoundError if the file is missing.
            with open(self.filename, "r") as file:
                content = file.read()
                if not content:
                    print("The journal is currently empty.")
                else:
                    print("\n--- YOUR JOURNAL ENTRIES ---")
                    print(content)
        except FileNotFoundError:
            print("Error: No journal file found. Try adding an entry first.")
        except Exception as e:
            print(f"An error occurred: {e}")

    def search_entries(self):
        """Searches for a keyword within the file."""
        keyword = input("Enter keyword or date (YYYY-MM-DD) to search for: ").lower()
        found = False
        try:
            with open(self.filename, "r") as file:
                # Splitting entries by the dashed line separator
                entries = file.read().split('-'*30)
                print(f"\n--- Search Results for '{keyword}' ---")
                for entry in entries:
                    if keyword in entry.lower():
                        print(entry.strip())
                        print('-'*30)
                        found = True
                
                if not found:
                    print("No matching entries found.")
        except FileNotFoundError:
            print("Error: Journal file does not exist.")

    def delete_all(self):
        """Deletes the journal file after confirmation."""
        if not os.path.exists(self.filename):
            print("No journal file exists to delete.")
            return

        confirm = input("Are you sure you want to delete all entries? (yes/no): ").lower()
        if confirm == 'yes':
            try:
                os.remove(self.filename)
                print("Journal deleted successfully.")
            except Exception as e:
                print(f"Error deleting file: {e}")
        else:
            print("Deletion cancelled.")

def main():
    manager = JournalManager()
    
    while True:
        print("\n--- Personal Journal Menu ---")
        print("1. Add a New Entry")
        print("2. View All Entries")
        print("3. Search for an Entry")
        print("4. Delete All Entries")
        print("5. Exit")
        
        choice = input("Select an option (1-5): ")
        
        if choice == '1':
            manager.add_entry()
        elif choice == '2':
            manager.view_entries()
        elif choice == '3':
            manager.search_entries()
        elif choice == '4':
            manager.delete_all()
        elif choice == '5':
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 5.")

# Run the program
if __name__ == "__main__":
    main()
```
